## Introduction
This is a movie data set containing important information about 10,000 movies collected from The Movie Database (TMDb). The dataset contain 10866 rows and 21 columns of information about the movies. This information in this data set is useful in understanding why and how movies become top sellers or a failure. Similar the dataset is useful in machine learning to predict the success of a  movie. 

### Dataset Description 

<br>
    1) id : A distinct identification number for each movie.

<br>

<br>
    10) tagline: A phrase or clause that present an idea from the movie.
<br>
    11) keywords: catchy words that shows the movie idea 
<br>
    12) overview: A short paragraph describing the movie.
<br>
    13) runtime: figure shwoing the length of the movie in minutes.
<br>
    14) genres: this column shows the categories a movie belong to based on type.

<br>

    19) release_year: The year the movie was released to public view

## Exploratory Data Analysis




#### Create new columns to strengthen analysis

```python
# 2 new columns named profit and profit_adj will be created to see how well each movie did
df['profit'] = df['revenue'] - df['budget']
df['profit_adj'] = df['revenue_adj'] - df['budget_adj']
df.head()
```

### Research Question 1: How many movies was released per year?

```python
#creating a pandas pivot table to count the number of movies released per year
movies_per_year = df.pivot_table(values = 'original_title', index = 'release_year', aggfunc = "count")


#confirming data
movies_per_year.head(70)
```

### Research Question 2: How much was grossed per year of production?

```python
#creating a pandas pivot table to sum the amount grossed per year
annual_gross = df.pivot_table(values = 'revenue', index = 'release_year', aggfunc = "sum")


annual_gross.head(70)
```

### Research Question 3: Which genre is the most produced?

```python
df['genres'].mode()
```

### Research Question 4: How much did highest grossing movie make?

```python
#using the max() function to get the highest grossing amount.

df['revenue'].max()
```

### Research Question 5: What are the most profitable movies?

```python
df.groupby('original_title').sum().revenue.nlargest(10).plot(kind='barh')
plt.title('Most profitable movies')
plt.xlabel('Profit')
plt.ylabel('Title')
plt.show()
```

### Research question 6: Which movies recieved the highest ratingd by votes?

```python
#average rating per director
df.groupby('original_title').mean().vote_average.nlargest(10).plot(kind='barh', color='red')
plt.title('Top movies by votes')
plt.xlabel('Average vote')
plt.ylabel('Title')
plt.show()

```
